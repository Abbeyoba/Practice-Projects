The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012 and subsequent reports have been presented annually till 2016. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th.  The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

```python
#Using Scatter plots, further analysis is done to show the relationship between happiness scores and other factors

#Happiness score vs other factors

tab_20_colors = ["#1f77b4", "#2ca02c", "#9467bd", "#e377c2", "#bcbd22", "#9edae5"]
fig, ax = plt.subplots(3, 2, figsize = (16,16))
i = 0
for j, column in enumerate(df_new[1:]):
    corr = df['Happiness Score'].corr(df[column])
    sns.regplot(df, x = "Happiness Score", y = column, color = tab_20_colors[i], ax = ax[j//2, j%2])
    ax[j//2, j%2].set_title("Correlation = {:.4f}".format(corr))
    i += 1 
fig.tight_layout()
fig.subplots_adjust(hspace = 0.3)
fig.show()
```

```python
#Using Boxplot, we can determine the happiness scores by region


fig = plt.figure(figsize = (10,7))
sns.boxplot(df, y = 'Region', x = "Happiness Score", palette = 'tab20')
plt.title("Happiness Score by Region", size = 20)
```

* Western Europe and North America Regions show higher levels of happiness and sub-saharan Africa with lowest as earlier shown

Conclusively, the analyses above have shown which countries of the world are most and least happy, and the regions that the countries in each category are present. The factors responsible for the level of happiness in each of these countries and the observed regions were also shown to include the higher GDP per capita, life expectancy, freedom, and generosity.

However, a negative correlation was observed between happiness score and trust of government on corruption and this was common among the countries in the sub-Saharan Africa with Togo being the least happy country. This implies that countries with high level of government corruption will have low levels of happiness among citizens
